```python
# INTENTO DE PROGRAMACIÓN EN PYTHON.
#Por Marcos_Pinon.

#Librería de espera por X segundos.
from time import sleep

 #Librería requerida para "round", al parecer jaja.
import math

#Librerías para visualización de imágenes. Créditos: https://www.youtube.com/watch?v=wiVhwDlECYc&t=408s
import cv2
from google.colab.patches import cv2_imshow

#Librerías para hacer las gráficas XD
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#Librería para cerrar programa.
import sys

#Librería para reproducción de sonidos. Créditos: https://stackoverflow.com/questions/68581187/playing-sound-in-google-colab
from IPython.display import Audio

#Páginas web para ingreso de colores predeterminados en mis gráficas:
#https://colorhunt.co/
#https://seaborn.pydata.org/examples/index.html Librería!!!


def juego_pal():
  sleep(4)
  print("""En esta ocasión, tenemos una bonita canción llamada "Ameno" de Era.\n\n""") #Puse triple comilla para que me lea las comillas del título de la canción.
  sleep(2)
  port = cv2.imread("/content/drive/MyDrive/Maestria/ameno.jpg")
  cv2_imshow(port)
  print("Por favor, espera mientras se descarga...\n\n")
  sleep(4)

  #Importación del texto para el ejercicio:
  # Fetch the dataset using the raw GitHub URL.
  !curl --remote-name \
      -H 'Accept: application/vnd.github.v3.raw' \
      --location https://raw.githubusercontent.com/MarcosPICM/tareas_progr/8a174a6c033e16fb659adb607ba49aff0091c6b7/ameno.txt #dirección donde se encuentra mi archivo...

  print("\nListo! Se ha descargado el archivo.\n\n")
  op_juego_pal() #PREGUNTAR por qué en este caso, si brincó directamente a una línea inferior.

def ret_line(): #Definí funciones para la lectura del archivo desde el inicio.
  with open("ameno.txt", "r", encoding='utf-8') as f: #Abre el archivo, y lo manda a un argumento "f".
    for line in f:
      print(line)
def ret_words():  #Definí funciones para la formación de diccionario desde el inicio.
  words = [] #Lista de palabras
  with open("ameno.txt", "r", encoding='utf-8') as f: #Abre el archivo, y lo manda a un argumento "f".
    for line in f:
      words.extend(line.replace(',', '').replace('.', '').replace('(','').replace(')','').lower().split()) #Lo modifiqué para que desde aquí depure las comas, puntos, y ponga todas en minusculas.
  print(words)

def depuracion_palabras():
  words = [] #Aquí repetí la misma orden de la función anterior (ret_words), solo que añadí el módulo de "depuración de palabras".
  with open("ameno.txt", "r", encoding='utf-8') as f:
    for line in f:
      words.extend(line.replace(',', '').replace('.', '').replace('(','').replace(')','').lower().split())
  #En esta línea, inicia el módulo de depuración.
  pal_comunes = ["lo", "para", "el", "de", "en", "la", "con", "cada", "tu", "tú", "una", "que", "te", "yo", "y", "por", "más", "a", "vas", "aunque", "me", "un", "no", "mi", "al", "son", "este", "así", "aquí", "o", "del", "he", "ese", ""]
  pal_depuradas = []
  for i in words: #Ciclo de iteración para cada palabra de la lista de la canción.
    if i not in pal_comunes: #Si en la canción no están las palabras comunes...
      pal_depuradas.append(i) #Aquí se van a guardar las palabras que no son conectores ni palabras comunes del texto.
  print(pal_depuradas)


def contadors():
  words = [] #Aquí repetí la misma orden de la función anterior (ret_words), solo que añadí el módulo de "depuración de palabras".
  with open("ameno.txt", "r", encoding='utf-8') as f:
    for line in f:
      words.extend(line.replace(',', '').replace('.', '').replace('(','').replace(')','').lower().split())
  pal_comunes = ["lo", "para", "el", "de", "en", "la", "con", "cada", "tu", "tú", "una", "que", "te", "yo", "y", "por", "más", "a", "vas", "aunque", "me", "un", "no", "mi", "al", "son", "este", "así", "aquí", "o", "del", "he", "ese"]
  pal_depuradas = [] #En la línea anterior, inicia el módulo de depuración
  for i in words:
    if i not in pal_comunes:
      pal_depuradas.append(i) #Aquí se van a guardar las palabras que no son conectores ni palabras comunes del texto.
  #Módulo de contador de palabras:
  contador = {}
  for k in pal_depuradas: #Bucle que itera a cada palabra...
    if k not in contador: #Si la palabra NO se encuentra en el contador... Al principio, pues no hay nada XD...
      contador[k] = 1  #cuando es la primera que encuentra, entonces le da el valor de "1". Y también va a agregar el dato al diccionario.
    else:
      contador[k] += 1 #Cuando ya hay más de una, entonces ya solamente aumenta el valor.
  return contador

def grafica():
  a=contadors() #Llamo al diccionario, colocándolo dentro de una variable "a".
  x=a.keys() #Defino los valores de X como las llaves ("keys") del diccionario.
  y=a.values() #Defino los valores de Y como los valores ("values") del diccionario.
  x1=[i for i in x] #Ciclo "for" integrado. Para construir una lista con las palabras. Es como si fuera el .append().
  y1=[i for i in y] #Ciclo "for" integrado. Para construir una lista con las frecuencias.
  #A partir de aquí, inicia el código para hacer la gráfica de barras.
  plt.bar(x1, y1, color=['blue','red','green','#8B322C','brown', '#4793AF', '#FFC470', '#DD5746'], edgecolor='none', capsize=5) #Método para construir la gráfica, donde se incluyen los valores en x, y, colores, ancho de los límites, etc.
  plt.xlabel("Letras", fontsize=15) #Etiquetas eje X, con tamaño de fuente.
  plt.ylabel("Frecuencia", fontsize=15) #Etiquetas eje Y, con tamaño de fuente.
  x_pos = range(len(x1)) # Para definir la cantidad de muestras.
  plt.xticks(x_pos, x1, rotation=90, fontsize=10) #Para modificar las etiquetas de cada valor en X.
  plt.show() #Comando para mostrar la gráfica.

  #Esto ya no lo usé, pero lo dejo para futuros usos :)
   #Llamo al diccionario.
  #pals = list(contador.keys()) #Se define una variable donde se guardan sólamente las palabras (determinado por keys ó llaves)
  #repiten = list(contador.values()) #Se define otra variable donde se guarda la cantidad de veces que se repite (determinado por los valores ó values)

  #plt.bar(range(len(contador)), repiten, tick_label=names) #
  #plt.show()

def op_juego_pal():
  opc_juego_pal = input("¿Qué deseas realizar con la canción?\nEscribe:\nA) Ver las Lyrics.\nB) Ver el top 10 de palabras que más se repiten en ella.\nC) Ver un gráfico del top 10 del inciso anterior.\n")
  juego_pal_list_A = ["A", "a"]
  juego_pal_list_B = ["B", "b"]
  juego_pal_list_C = ["C", "c"]

  if opc_juego_pal in juego_pal_list_A:
    print(f"Ok. Elegiste {opc_juego_pal}.\nAquí la tienes.\n\nDisfrútala:\n\n")
    sleep(3)
    #global texto_entero
    #texto_entero = open("ameno.txt", "r").read() # Open, para abrir archivos. "r" = Abrir en modo lectura.
    #print(f"Aquí tienes: {texto_entero}")
    reproductor = Audio("/content/drive/MyDrive/Maestria/Ameno.mp3", autoplay=True)
    display(reproductor)
    ret_line()


  elif opc_juego_pal in juego_pal_list_B:
    print(f"Ok. Elegiste {opc_juego_pal}.\n\nAquí te muestro las palabras originales en forma de lista.\n\n")
    sleep(3)
    ret_words() #Lista original.
    sleep(2)
    print("\n\nA continuación, te muestro la lista de palabras que no se repiten:\n\n")
    sleep(2)
    depuracion_palabras() #Lista depurada.
    print("\n\nAhora sí... Te muestro la cantidad de veces que se repiten las palabras...\n\n")
    sleep(2)
    print(contadors()) #Devuelve el contador de palabras.

  elif opc_juego_pal in juego_pal_list_C:
    print(f"Ok. Elegiste {opc_juego_pal}.\n\nA continuación, te mostraré un gráfico de barras con la frecuencia con la que se repiten las palabras de la canción :)\n\n")
    sleep(4)
    grafica()

  else:
    while opc_juego_pal != juego_pal_list_A or juego_pal_list_B or juego_pal_list_C:
      print("Lo siento. No te entendí. Por favor, intenta de nuevo.\n")
      op_juego_pal()



def calc_list_A(calc): #Para transformar de aM a cp/μL.
    print(f"¡De acuerdo! Elegiste {calc}.") #Este es el enlace para que pueda llamar esta función.
    dato_conc_prim = float(input("¿Cuál es la concentración inicial?\n Recuerda que tiene que ser en aM.\n")) #Ingresa el usuario el valor. Define el valor como "float" (número decimal).
    dato_conc_prim = round(dato_conc_prim, 3) #Función para redondear el valor ingresado, de ser necesario.
    print(f"Ok. {dato_conc_prim} aM equivale a...") #Imprime el valor ingresado. Ya redondeado a 2 decimales.
    def conv_prim(dato_conc_prim): #Definiendo la operación, en función del valor de concentración ingresado por el usuario.
       return dato_conc_prim * (1e-18) * (6.022e23) * (1e-6) #Función "return" de declaración, para que devuelva el valor.
    sleep(3)
    print(f"{float(round(conv_prim(dato_conc_prim), 3))} cp/μL.") #Aquí para imprimir, incluyo todo lo necesario para redondear y definir como decimal.

def calc_list_B(calc): #Funciona bajo el mismo principio que lo que mencioné en el anterior ejemplo.
    print(f"¡De acuerdo! Elegiste {calc}.")
    dato_conc_seg = float(input("¿Cuál es la concentración inicial?\nRecuerda que tiene que ser en cp/μL.\n"))
    dato_conc_seg = round(dato_conc_seg, 3)
    print(f"Ok. {dato_conc_seg} cp/μL equivale a...")
    def conv_seg(dato_conc_seg):
       return dato_conc_seg * (1e18) * ((1)/(6.022e23)) * (1e6)
    sleep(3)
    print(f"{float(round(conv_seg(dato_conc_seg), 3))} aM.")


#Para la calculadora, haré una de conversiones que me ayuda en mi práctica diaria (hago un biosensor y mis concentraciones son muy bajas jaja)
def conversiones():
  sleep(2)
  print("Te iré solicitando unos datos necesarios para realizar las conversiones de concentración.\n")
  sleep(2)
  print("Recuerda que esta calculadora te ayuda con concentraciones bajas")
  sleep(2)
  calc = input("¿Qué tipo de conversión requieres?\n Teclea:\n A) aM a cp/μL;\n B) cp/μL a aM\n")
  calc_list_prim = ["a", "A"]
  calc_list_seg = ["b", "B"]

  if calc in calc_list_prim:
    calc_list_A(calc) #Para que pueda llamar a la siguiente función, la coloqué en función del argumento "calc".

  elif calc in calc_list_seg:
    calc_list_B(calc)

  else:
    while calc != calc_list_prim or calc_list_seg:
      print("Lo siento. No te entendí. Por favor, intenta de nuevo.\n")
      conversiones() #Función recursiva.


def inicio():
  preg_inicio = input("Qué te gustaría realizar el día de hoy?\nEscribe:\n A) Apoyo con tus conversiones de concentración.\n B) Realizar un juego de palabras.\n C) Salir del programa.\n")
  inicio_A = ["A", "a"] #Lista con las opciones con letras capitalizadas y no capitalizadas.
  inicio_B = ["B", "b"]
  inicio_C = ["C", "c"]

  while True:
    if preg_inicio in inicio_A:
      print(f"Perfecto. Elegiste {preg_inicio}:\n\nApoyo con conversiones.\n\n") #PREGUNTAR cómo funciona "f". Y por qué se pone entre llaves el resultado de lo ingresado por el usuario (sintaxis).
      conversiones() #Manda a bloque/función de conversiones.
      break
      #f indica que quiero que imprima una variable que va entre las llaves.También funciona para números. Indica que quiere que imprimamos el resultado de una variable.

    elif preg_inicio in inicio_B:
      print(f"Perfecto. Elegiste {preg_inicio}:\n\nJuego de palabras.\n\n")
      juego_pal()
      break

    elif preg_inicio in inicio_C: #Falta hacerla. Con los 2 puntos estamos diciendo que haga lo siguiente.:
      print("Muy bien.\n\nMuchas gracias por ingresar a este programa!\n\nY recuerda...\n\n")
      sleep(2)
      print("Hit the target\n\n")
      print("""
                                                            |
                                                            \.
                                                            /|.
                                                          /  `|.
                                                        /     |.
                                                      /       |.
                                                    /         `|.
                                                  /            |.
                                                /              |.
                                              /                |.
          __                                 /                  `|.
          -\                              /                     |.
            \\                          /                       |.
              \\                      /                         |.
                \|                   /                           |\
                  \#####\          /                             ||
              ==###########>     /                               ||
              \##==      \    /                                 ||
          ______ =       =|__/___                                ||
      ,--' ,----`-,__ ___/'  --,-`-==============================##==========>
    \               '        ##_______ ______   ______,--,____,=##,__
      `,    __==    ___,-,__,--'#'  ==='      `-'              | ##,-/
        `-,____,---'       \####\              |        ____,--\_##,/
            #_              |##   \  _____,---==,__,---'         ##
            #              ]===--==\                            ||
            #,             ]         \                          ||
              #_            |           \                        ||
              ##_       __/'             \                      ||
                ####='     |                \                    |/
                ###       |                  \                  |.
                ##       _'                    \                |.
                ###=======]                       \              |.
              ///        |                         \           ,|.
              //         |                           \         |.
                                                        \      ,|.
                                                          \    |.
                                                            \  |.
                                                              \|.
                                                              /.
                                                            |
    """)
      sleep(2)
      break
      #También podría usar sys.exit() PREGUNTAR
      #sys.exit()

    else: #La palabra reservada está esperando la indicación anterior.
      while preg_inicio != inicio_A or inicio_B or inicio_C: #Bucle para... Si la respuesta del usuario es distinta a las establecidas en las listas...
        print("Lo siento. No te entendí. Por favor, intenta de nuevo.\n") #Regresa mensaje de error :(
        sleep(2)
        inicio() #Vuelve al inicio de este bloque o función...

#Inicio del programa
print("Hola! Bienvenido a esta plataforma de entretenimiento\nDonde podrás echar un poco de procrastinación productiva XD.\n")
inicio()
#Aquí aprendí que al parecer los "bloques" (funciones) se colocan al INICIO, y no después de mi proceso inicial. Es decir, todo al reves :(

```
